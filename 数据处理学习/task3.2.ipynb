```python
import matplotlib.pyplot as plt
# %matplotlib inline 

from matplotlib_inline import backend_inline   
backend_inline.set_matplotlib_formats('svg') 

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']   
plt.rcParams['axes.unicode_minus'] = False  #支持负号显示
import numpy as np
L0 = 1
L1 = 3

k0 = 1
k1 = 5

x0 = 0
x1 = 6

x = np.linspace(-10,10,200)

y0 = L0/(1+np.exp(k0*(x0-x)))
y1 = L1/(1+np.exp(k0*(x0-x)))  #只改变L，为曲线饱和值，图像变陡，最高峰值变大
y2 = L0/(1+np.exp(k1*(x0-x)))  #只改变k，为曲线陡峭度，一开始增长较慢，到图像中间上升部分变陡，但最高峰值不变，
y3 = L0/(1+np.exp(k0*(x1-x)))  #只改变x0，为曲线的中点位置，x0变大，图像整体向右平移
y4 = L1/(1+np.exp(k1*(x1-x)))  #综合改变L、k、x0的效果


fig1,ax1 = plt.subplots()
ax1.plot(x,y0,linewidth='2',label = 'L=1,k=1,x0=0')
ax1.plot(x,y1,linewidth='2',label = 'L=3,k=1,x0=0')
ax1.plot(x,y2,linewidth='2',label = 'L=1,k=5,x0=0')
ax1.plot(x,y3,linewidth='2',label = 'L=1,k=1,x0=6')
ax1.plot(x,y4,linewidth='2',label = 'L=3,k=5,x0=6')

ax1.legend(loc='upper left',frameon = False, ncol=1)
ax1.set_title('Logistic函数曲线')
ax1.set_xticks(np.arange(-10,11,1))
ax1.set_yticks(np.arange(0,3.6,0.4))


delt_L = np.linspace(-5,5,10)
delt_k = np.linspace(-2,8,10)
delt_x0 = np.linspace(-5,5,10)

fig2,ax2 = plt.subplots(3,figsize=(10, 15))  #探究同一参数改变对函数图像影响程度
ax2[0].set_title('改变L值对Logistic函数曲线的影响')
ax2[1].set_title('改变k值对Logistic函数曲线的影响')
ax2[2].set_title('改变x0值对Logistic函数曲线的影响')

for i in delt_L:
    y5 = i/(1+np.exp(k0*(x0-x)))   #峰值正负影响突起方向，图像呈花洒状
    ax2[0].plot(x,y5)

for i in delt_k:
    y5 = L0/(1+np.exp(i*(x0-x)))  #改变陡峭度，图像呈扇形状，负数时从峰值下落，为0时曲线近似线性，总体呈X型
    ax2[1].plot(x,y5)

for i in delt_x0:
    y5 = L0/(1+np.exp(k0*(i-x)))  #改变中点位置，图像水平平移，呈彩带状
    ax2[2].plot(x,y5)
```
